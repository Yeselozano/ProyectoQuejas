```python
## Creamos un nuevo dataframe con las principales solicitudes
DF_Quejas_Comisarias_Gen = DF_Quejas_Comisarias.groupby("GENERO").size(). sort_values(ascending= False).head(20)
DF_Quejas_Comisarias_Gen
```

```python
"""Histrograma: es una gráfica donde primero se ordenan de menor a mayor los registros y luego se cuentan cada uno de ellos"""

##Histograma por género

plt.figure(figsize=(4,2))                     ## Se establece el tamaño del gráfico
DF_Quejas_Comisarias ["GENERO"].hist()          ## Se llama el campo a graficar
plt.title("Histograma de Solicitudes por Género")             ## Se establece título para el gráfico
plt.xlabel("Género")                          ## Se establece título para el eje x
plt.ylabel("Cantidad en miles")                      ## Se establece título para el eje y
plt.show()                                    ## Muestra la gráfica
```

```python
## Gráfico  de una torta para distribución de porcentaje en Solicitudes de la Of. de Atención a usuarios de Comisarías de Familia
grafico_tortas_plot = DF_Quejas_Comisarias_Gen.plot.pie(
    figsize=(8,8),
    autopct="%.2f")
plt.title("Distribución porcentual solicitudes por género", bbox={'facecolor':"0.8",'pad':4})
plt.show()
```

```python
"""AUTORECONOCIMIENTO ETNICO-RACIAL"""
## Creamos un nuevo dataframe con usuarios por Autoreconocimiento Etnico Racial
DF_Quejas_Comisarias_Rac = DF_Quejas_Comisarias.groupby("AUTORECONOCIMIENTO ETNICO-RACIAL").size(). sort_values(ascending= False).head(20)
DF_Quejas_Comisarias_Rac
```

```python
"""PERSONAS DE ESPECIAL INTERES"""
## Creamos un nuevo dataframe con usuarios por Autoreconocimiento Etnico Racial
DF_Quejas_Comisarias_Esp = DF_Quejas_Comisarias.groupby("PERSONAS DE ESPECIAL INTERES").size(). sort_values(ascending= False).head(20)
DF_Quejas_Comisarias_Esp
```

```python
##Histograma por AUTORECONOCIMIENTO ETNICO-RACIAL

plt.figure(figsize=(4,2))                     ## Se establece el tamaño del gráfico
DF_Quejas_Comisarias ["AUTORECONOCIMIENTO ETNICO-RACIAL"].hist()  ## Se llama el campo a graficar
plt.title("Histograma de Solicitudes por AUTORECONOCIMIENTO ETNICO-RACIAL") ## Se establece título para el gráfico
plt.xticks(rotation=90)                       ## Roto los títulos del eje x
plt.xlabel("AUTORECONOCIMIENTO ETNICO-RACIAL")## Se establece título para el eje x
plt.ylabel("Cantidad en miles")               ## Se establece título para el eje y
plt.show()                                    ## Muestra la gráfica
```

```python
## Gráfico  de una torta para distribución de porcentaje en Solicitudes de la Of. de Atención a usuarios de Comisarías de Familia
grafico_tortas_plot = DF_Quejas_Comisarias_Rac.plot.pie(
    figsize=(8,8),
    autopct="%.1f")
plt.title("Distribución porcentual solicitudes por AUTORECONOCIMIENTO ETNICO-RACIAL", bbox={'facecolor':"0.8",'pad':4})
plt.show()
```

```python
## Creamos un nuevo dataframe con las principales solicitudes
DF_Quejas_Comisarias_Sol = DF_Quejas_Comisarias.groupby("TIPO DE SOLICITUD").size(). sort_values(ascending= False).head(20)
DF_Quejas_Comisarias_Sol
```

```python
#Histograma de Solicitudes

# plt.figure(figsize=(100,1))       ## Creo el tamaño de la gráfica
plt.hist(DF_Quejas_Comisarias_Sol.index, weights=DF_Quejas_Comisarias_Sol.values) ## Tomo los índices para el eje x y para el eje y las cantidades
plt.xticks(rotation=90)         ## Roto los títulos del eje x
plt.xlabel("TIPO DE SOLICITUD") ## Título para el eje x
plt.ylabel("CANTIDADES")        ## Título para el eje y
plt.title("Histograma de Solicitudes - Oficina de Atención a Usuarios de Comisarías de Familia")   ## Título para la gráfica
plt.show()
```

```python
## Gráfico  de una torta para distribución de porcentaje en Solicitudes de la Of. de Atención a usuarios de Comisarías de Familia
grafico_tortas_plot = DF_Quejas_Comisarias_Sol.plot.pie(
    figsize=(8,8),
    autopct="%.2f")
plt.title("Distribución porcentual del tipo de solicitud", bbox={'facecolor':"0.8",'pad':4})
plt.show()
```
